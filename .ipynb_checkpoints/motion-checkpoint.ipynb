````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import subprocess
%matplotlib notebook
````

# Motion Profile Visualization
This uses our C++ robot code to output motion profile points to a `.csv` file we use to plot.

Units are:

- time = ms
- distance = encoder ticks (plot_profile called with inches for convenience)
- velocity = ticks/100ms
- acceleration = ticks/(100ms)<sup>2</sup>

Usage of the profile program to generate data:

```
$ motion <time full vel (ms)> <time full accel (ms)> <max speed (ticks/100ms)> <travel distance (ticks)>
```

````python
def plot_profile(v_t, a_t, v_max, d):
    ticks = d * 1983.65 # ticks per inch
    subprocess.run("./build/install/motion/bin/motion {} {} {} {} {}".format(v_t, a_t, v_max, ticks, 10), shell=True)
    df = pd.read_csv("output.csv", index_col="Time (ms)")
    df.plot()
````

Distance to travel 2347 ticks (4 ft.), max velocity 587 ticks/100ms (10 ft/sec), time to max velocity 200ms, time to max acceleration 100ms.

````python
plot_profile(200, 100, 587, 82)
````
